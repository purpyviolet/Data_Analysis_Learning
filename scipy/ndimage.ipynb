### **图像处理模块 (`scipy.ndimage`)**

`scipy.ndimage` 模块提供了一些高效的多维图像处理功能，包括图像变换、滤波和形态学操作。


### **1. 图像变换**
- **平移**: `shift` 函数。
- **旋转**: `rotate` 函数。
- **缩放**: `zoom` 函数。

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import shift, rotate, zoom

# 创建示例图像
image = np.zeros((100, 100))
image[30:70, 30:70] = 1

# 平移
shifted_image = shift(image, shift=(10, 20))

# 旋转
rotated_image = rotate(image, angle=45, reshape=False)

# 缩放
zoomed_image = zoom(image, zoom=1.5)

# 可视化
plt.figure(figsize=(12, 4))
plt.subplot(1, 4, 1)
plt.title("Original Image")
plt.imshow(image, cmap='gray')

plt.subplot(1, 4, 2)
plt.title("Shifted Image")
plt.imshow(shifted_image, cmap='gray')

plt.subplot(1, 4, 3)
plt.title("Rotated Image")
plt.imshow(rotated_image, cmap='gray')

plt.subplot(1, 4, 4)
plt.title("Zoomed Image")
plt.imshow(zoomed_image, cmap='gray')

plt.tight_layout()
plt.show()
```

### **2. 图像过滤**
- **高斯滤波**: 用于平滑图像并去除噪声。
- **中值滤波**: 用于去除椒盐噪声。


```python
from scipy.ndimage import gaussian_filter, median_filter

# 添加噪声的示例图像
noisy_image = image + 0.4 * np.random.random(image.shape)

# 高斯滤波
gaussian_filtered_image = gaussian_filter(noisy_image, sigma=2)

# 中值滤波
median_filtered_image = median_filter(noisy_image, size=3)

# 可视化
plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
plt.title("Noisy Image")
plt.imshow(noisy_image, cmap='gray')

plt.subplot(1, 3, 2)
plt.title("Gaussian Filter")
plt.imshow(gaussian_filtered_image, cmap='gray')

plt.subplot(1, 3, 3)
plt.title("Median Filter")
plt.imshow(median_filtered_image, cmap='gray')

plt.tight_layout()
plt.show()
```

### **3. 形态学操作**
- **腐蚀**: `binary_erosion`。
- **膨胀**: `binary_dilation`。
- **开运算**: `binary_opening`。
- **闭运算**: `binary_closing`。

```python
from scipy.ndimage import binary_erosion, binary_dilation, binary_opening, binary_closing

# 创建二值图像
binary_image = np.zeros((100, 100))
binary_image[30:70, 30:70] = 1

# 腐蚀
eroded_image = binary_erosion(binary_image)

# 膨胀
dilated_image = binary_dilation(binary_image)

# 开运算
opened_image = binary_opening(binary_image)

# 闭运算
closed_image = binary_closing(binary_image)

# 可视化
plt.figure(figsize=(12, 8))
plt.subplot(2, 2, 1)
plt.title("Original Image")
plt.imshow(binary_image, cmap='gray')

plt.subplot(2, 2, 2)
plt.title("Erosion")
plt.imshow(eroded_image, cmap='gray')

plt.subplot(2, 2, 3)
plt.title("Dilation")
plt.imshow(dilated_image, cmap='gray')

plt.subplot(2, 2, 4)
plt.title("Opening and Closing")
plt.imshow(opened_image + closed_image, cmap='gray')

plt.tight_layout()
plt.show()
```

### **总结**

| **功能**       | **方法**                      | **作用**                                             |
|----------------|--------------------------------|-----------------------------------------------------|
| **图像变换**   | `shift`, `rotate`, `zoom`     | 平移、旋转、缩放图像。                               |
| **图像过滤**   | `gaussian_filter`, `median_filter` | 平滑图像、去噪声（高斯和中值滤波）。                |
| **形态学操作** | `binary_erosion`, `binary_dilation`, `binary_opening`, `binary_closing` | 图像形态学处理：腐蚀、膨胀、开运算、闭运算。 |
