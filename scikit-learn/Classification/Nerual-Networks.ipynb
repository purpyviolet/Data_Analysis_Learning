```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
from sklearn.neural_network import MLPClassifier

# Generate synthetic dataset for classification
X, y = make_classification(
    n_samples=1000,
    n_features=20,
    n_classes=2,
    random_state=42
)

# Split data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# Define the MLPClassifier model
mlp_clf = MLPClassifier(
    hidden_layer_sizes=(100,),  # One hidden layer with 100 neurons
    activation='relu',          # Activation function
    solver='adam',              # Optimizer
    max_iter=500,               # Maximum number of iterations
    random_state=42             # Random seed
)

# Train the model
mlp_clf.fit(X_train, y_train)

# Make predictions
y_pred = mlp_clf.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f"MLPClassifier Accuracy: {accuracy:.4f}")
print("\nClassification Report:")
print(classification_report(y_test, y_pred))

# Visualize weights of the first layer
plt.figure(figsize=(10, 6))
plt.title("First Layer Weights (Heatmap)")
plt.imshow(mlp_clf.coefs_[0], cmap='coolwarm', aspect='auto')
plt.colorbar(label='Weight Value')
plt.xlabel("Hidden Neurons")
plt.ylabel("Input Features")
plt.show()
```
