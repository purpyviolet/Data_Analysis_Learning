```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
from sklearn.neural_network import MLPClassifier

# Generate synthetic dataset for classification
X, y = make_classification(
    n_samples=1000,
    n_features=20,
    n_classes=2,
    random_state=42
)

# Split data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# Define the MLPClassifier model
mlp_clf = MLPClassifier(
    hidden_layer_sizes=(100,),  # One hidden layer with 100 neurons
    activation='relu',          # Activation function
    solver='adam',              # Optimizer
    max_iter=500,               # Maximum number of iterations
    random_state=42             # Random seed
)

# Train the model
mlp_clf.fit(X_train, y_train)

# Make predictions
y_pred = mlp_clf.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f"MLPClassifier Accuracy: {accuracy:.4f}")
print("\nClassification Report:")
print(classification_report(y_test, y_pred))

# Visualize weights of the first layer
plt.figure(figsize=(10, 6))
plt.title("First Layer Weights (Heatmap)")
plt.imshow(mlp_clf.coefs_[0], cmap='coolwarm', aspect='auto')
plt.colorbar(label='Weight Value')
plt.xlabel("Hidden Neurons")
plt.ylabel("Input Features")
plt.show()
```



### **MLPClassifier 的特点、适用场景和优缺点**

#### **特点**
- 多层感知机分类器，支持自定义隐藏层、激活函数和优化器。
- 支持多分类和二分类任务。
- 使用反向传播算法进行优化。

#### **适用场景**
- 中小规模的分类任务。
- 非线性分类任务，例如基于复杂模式的分类。
- 数据集较小且不需要深度学习框架时。

#### **优点**
- 易于使用，参数可调节性强。
- 支持多分类任务。
- 适合非线性数据的分类。

#### **缺点**
- 对参数敏感（如隐藏层大小、激活函数、优化器等）。
- 训练时间可能较长，特别是当数据规模较大时。
- 相较于深度学习框架（如 TensorFlow、PyTorch），不适合大规模深度网络的构建。
